**Q.1 <!DOCTYPE html> is it a tag of html? If not, what is it and why do we use it?**

**ANS**
Yes, <!DOCTYPE html> is indeed a tag in HTML. It is a special instruction or declaration that tells the web browser which version of HTML the web page is written in.

Think of <!DOCTYPE html> as a way of saying to the web browser, "Hey, this web page is using the latest version of HTML called HTML5." It helps the browser understand how to read and display the content of the web page correctly.

Using the correct <!DOCTYPE> declaration is important because different versions of HTML have different rules and features. By specifying the correct <!DOCTYPE>, web developers ensure that their web pages are interpreted and displayed correctly by different web browsers.

In simpler terms, <!DOCTYPE html> is like a little note at the beginning of an HTML document that helps the browser understand and show the web page properly. It's like telling the browser, "Hey, this is the type of HTML we're using, so please display it correctly."

**Q.2 Explain Semantic tags in html? And why do we need it?**

**ANS** Semantic tags in HTML are special tags that give meaning and structure to the content of a web page. They describe the type of content they wrap, such as headers, paragraphs, lists, articles, sections, and more. 

We need semantic tags because they help search engines and assistive technologies understand the content better. It improves accessibility, SEO, and makes it easier for developers to style and organize the web page. Semantic tags make the HTML code more meaningful and easier to maintain.

**Q.3 Differentiate between HTML Tags and Elements?**

**ANS**
HTML Tags and Elements are related but have different meanings:

HTML Tags are the keywords or names surrounded by angle brackets (< and >) that define the structure and functionality of specific elements. For example, `<h1>` is a tag used to define a heading.

HTML Elements, on the other hand, consist of the opening and closing tags, along with the content within them. Elements are made up of tags and the content they enclose, like `<h1>Welcome!</h1>`, where `<h1>` is the opening tag, "Welcome!" is the content, and `</h1>` is the closing tag.

**Q.6 What are some of the advantages of HTML5 over its previous versions?**

HTML5 brings several advantages over its previous versions:

1. Improved multimedia support: HTML5 allows embedding audio and video directly without relying on external plugins like Flash.

2. Enhanced form handling: HTML5 introduces new form input types and attributes, making it easier to create user-friendly and interactive forms.

3. Mobile-friendly: HTML5 provides better support for mobile devices, allowing responsive design and native-like app experiences.

4. Offline capabilities: HTML5 enables web applications to work offline, providing better user experiences and reducing reliance on constant internet connectivity.

5. Simplified coding: HTML5 introduces cleaner and more semantic code structure, making it easier for developers to understand, maintain, and optimize their websites.

**Q.8 What is the difference between figure tag and img tag?**

The `<figure>` tag and the `<img>` tag serve different purposes:

- The `<img>` tag is used to display an image on a web page. It defines the source, size, and alternative text for the image.

- The `<figure>` tag, on the other hand, is used to group and caption a self-contained content, such as an image, illustration, or video. It provides a semantic structure for the content, allowing the caption to be associated with the figure.

In summary, `<img>` is for displaying an image, while `<figure>` is for organizing and captioning a self-contained content.

**Q.9 What’s the difference between html tag and attribute and give example of some global attributes?**


- HTML tags define the structure and content of elements on a web page, such as headings, paragraphs, and images.
- Attributes provide additional information or properties to the HTML elements. They modify the behavior or appearance of elements. For example, the "src" attribute specifies the source of an image.

Some examples of global attributes that can be used with multiple HTML tags are:
- "class" for defining a class or a group.
- "id" for providing a unique identifier.
- "style" for adding inline styles.
- "title" for displaying a tooltip.
- "data-*" for storing custom data.
